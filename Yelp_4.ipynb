```python
import networkx as nx
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data4 = ['business', 'city', 'b_city']
```

```python
df4 = pd.read_csv('/Users/charanyaponnala/Desktop/GSU/GRA/Truist Sprint/Yelp/business_city.txt', names = data4, delim_whitespace=True)
```

```python
df4.head()
```

```python
b_city_graph = nx.from_pandas_edgelist(df4, source = 'business', target = 'city', edge_attr = 'b_city', create_using = nx.Graph)
```

```python
type(b_city_graph)
```

```python
b_city_graph.nodes()
```

```python
df4['business'].nunique()
```

```python
plt.figure(figsize = (15,15))
nx.draw(b_city_graph, with_labels = True, arrows = True, node_color = 'g', edge_color = 'r',
                node_size = 1100, connectionstyle='arc3, rad = 0.1')
```

```python
node1 = df4['business'] # node list
node2 = df4['city']
edge_cols = ['b_city']
bcity_graph = nx.convert_matrix.from_pandas_edgelist(df4,
                                               source = 'business',
                                               target = 'city',
                                               edge_attr = edge_cols)
```

```python
fig = plt.figure(figsize = (15,15))
graph = nx.les_miserables_graph()
nx.draw(bcity_graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (15,15))
graph = nx.les_miserables_graph()
nx.draw(bcity_graph, 
        node_color = 'C1',
        node_size = 30,
        edge_color = 'g')
fig.set_facecolor('k')
```
